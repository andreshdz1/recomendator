```python
import pandas as pd

df = pd.read_csv('../data/recipes.csv')
df.head()
```

### Preprocesing 


```python
df = df.drop(['serving_size', 'tags'], axis=1)
df.head()
```

```python
df.info()
```

```python
df = df.fillna('unknown')
df.info()
```

```python
df['search_terms'] = df['search_terms'].str.strip('{}').str.replace("'","").str.replace(" ","").str.split(',')
df['ingredients'] = df['ingredients'].str.strip('[]').str.replace("'","").str.split(',')
df.head()
```

```python
df['search_terms'][1][0]
```

```python
from collections import Counter

all_words = [word for sublist in df['search_terms'] for word in sublist]
word_counts = Counter(all_words)
repeated_words = {word: count for word, count in word_counts.items() if count > 1}
repeated_words
```

#### lets distribute the chamba

```python


from pyspark.sql import SparkSession
from pyspark.ml.feature import OneHotEncoder, StringIndexer
from pyspark.ml import Pipeline
from pyspark.sql import SparkSession
from pyspark.ml.feature import OneHotEncoder, StringIndexer
from pyspark.ml import Pipeline
from pyspark.sql.functions import explode, col, collect_list

spark = SparkSession.builder.appName('recipes').getOrCreate()
## Convert the pandas DataFrame to a Spark DataFrame
df_spark = spark.createDataFrame(df)
## and register it as a temporary table because each entry is a list
df_exploded = df_spark.withColumn("ingredient", explode(col("ingredients"))) \
                      .withColumn("search_term", explode(col("search_terms")))

indexers = [StringIndexer(inputCol=column, outputCol=f"{column}_index") 
            for column in ["ingredient", "search_term"]]

encoders = [OneHotEncoder(inputCol=f"{column}_index", outputCol=f"{column}_vec") 
            for column in ["ingredient", "search_term"]]

pipeline = Pipeline(stages=indexers + encoders)
df_transformed = pipeline.fit(df_exploded).transform(df_exploded)

# Aggregate the results 
df_aggregated = df_transformed.groupBy("id").agg(
    collect_list("ingredient_vec").alias("ingredients_vec"),
    collect_list("search_term_vec").alias("search_terms_vec")
)
from pyspark.sql.functions import explode, col, collect_list

spark = SparkSession.builder.appName('recipes').getOrCreate()
## Convert the pandas DataFrame to a Spark DataFrame
df_spark = spark.createDataFrame(df)
## temporary table because each entry is a list
df_exploded = df_spark.withColumn("ingredient", explode(col("ingredients"))) \
                      .withColumn("search_term", explode(col("search_terms")))

indexers = [StringIndexer(inputCol=column, outputCol=f"{column}_index") 
            for column in ["ingredient", "search_term"]]

encoders = [OneHotEncoder(inputCol=f"{column}_index", outputCol=f"{column}_vec") 
            for column in ["ingredient", "search_term"]]

pipeline = Pipeline(stages=indexers + encoders)
df_transformed = pipeline.fit(df_exploded).transform(df_exploded)

# Aggregate the results back to the original structure
df_aggregated = df_transformed.groupBy("id").agg(
    collect_list("ingredient_vec").alias("ingredients_vec"),
    collect_list("search_term_vec").alias("search_terms_vec")
)
```
